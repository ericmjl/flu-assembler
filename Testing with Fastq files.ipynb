```python
# Find the minimum length
min_length = np.inf
len_counts = Counter()
for s in SeqIO.parse(fastq, 'fastq'):
    if len(s.seq) < min_length:
        min_length = len(s.seq)
    len_counts[len(s.seq)] += 1
```
